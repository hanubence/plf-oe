```python
def get_previousdates_rawsensor (dataframe_name, outcome_variable, sensor_variable, sensor_time_aggregation, N_days_back):
    
    '''Input:
    dataframe_name: the name of the dataframe
    sensor_variable: the name of the behavior measured witht the sensor
    sensor_time_aggregation: how often are we aggregating the sensor data (every 3, 6, 12, or 24 hours)
    N_days_back: how long back I want to go in days. Number of days will be transformed into time units.
                Number of time units depend on the sensor_time_aggregation.
                i.e. for sensor data aggregated every 6 hours, 1 day will have 4 units (24hrs / 6hrs)
                4 units * number of days I want to go back 14 = 4 * 14 = 56. I will create a list of numbers
                from 1 to 56.
                
    Output:
    a dataframe that is directly stored in my computer where each row is the date_time assigned to the outcome,
    and columns correspoding to the sensor values at the different time_back units.
    '''
    
    df_subset = dataframe_name[['cow_id', 'datetime_stamp', outcome_variable]]
    df_shift = dataframe_name.loc[(dataframe_name[sensor_variable].notnull())][['cow_id',sensor_variable]]
    
    # Create a list with numbers corresponding to time_back units
    time_aggregation = sensor_time_aggregation
    N_days = N_days_back
    num_timeback = (24/time_aggregation) * N_days
    my_list = range(1, num_timeback + 1)
    
    
    # Get the previous data point
    for i in my_list:
        df_shift['t_minus' + str(i)] = df_shift.groupby('cow_id')[sensor_variable].apply(lambda x: x.shift(i))

    
    # Merge the new dataframe with values shifted with the orginal one with the outcome
    # Keep only those rows where there is outcome
    result = pd.merge(df_subset, df_shift, left_index=True, right_index = True, how='outer', indicator = True)    
    result = result.loc[(result[outcome_variable].notnull())]
   

    # Generate files
    path_to_file = '/Users/gemavidal/Box Sync/Vet/Projects and Papers/2021 Chapter 4 Dissertation High SCC Prediciton Udder Hygiene/df_' + sensor_variable + '.csv'
    result.to_csv(path_to_file, index = None, header = True)
    
    
    return path_to_file
```

```python
variables = ['rumination_diff_6Hmean','rumination_diff_6Hvar', 'rumination_diff_6Hsum',
             'eating_diff_6Hmean','eating_diff_6Hvar', 'eating_diff_6Hsum',
             'notactive_diff_6Hmean','notactive_diff_6Hvar', 'notactive_diff_6Hsum',
             'active_diff_6Hmean','active_diff_6Hvar', 'active_diff_6Hsum',
             'highactivity_diff_6Hmean','highactivity_diff_6Hvar', 'highactivity_diff_6Hsum',
             'temperature_diff_6Hmean', 'temperature_diff_6Hvar', 'temperature_diff_6Hsum']

for i in variables:
    get_previousdates_rawsensor(df, 'scc_catg', i, 6, 14)
```
